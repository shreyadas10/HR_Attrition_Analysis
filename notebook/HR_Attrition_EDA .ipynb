```python
#business travel column insight
plt.figure(figsize=(10,5))
ax = sns.countplot(data=df,  x='BusinessTravel', hue='Attrition')

total_per_group = df['BusinessTravel'].value_counts()

for container in ax.containers:
    labels = []
    for i ,bar in enumerate(container):
        count = bar.get_height()
        total = total_per_group[i//2]
        labels.append(f'{count/total*100:.1f}%')
    ax.bar_label(container,labels=labels, fontsize=11)

plt.title("Attrition by Business Travel (%)")
plt.ylabel("Percentage of Employees")
plt.show()
```

```python
df.groupby('BusinessTravel')['Attrition'].value_counts(normalize=True)*100
```

- Insight:-
- Employees who have traveled frequently have higher Attrition rates than those who have travelled rarely or not at all,  indicating that frequent travelling can be one of the major reasons of leaving the organization.

```python
#mapping of Education column
education_mapping = {
    1: 'Below College',
    2: 'College',
    3: 'Bachelor',
    4: 'Master',
    5: 'Doctor'
}

#creating new column 
df['Education_level'] = df['Education'].map(education_mapping)


#dropping old column 
df.drop(columns =['Education'], inplace=True)
```

```python
#plotting attrition rate(%) by education on barplot

#calculating attrition rate
edu_attrition = df.groupby('Education_level')['Attrition'].value_counts(normalize=True).unstack() * 100

#attrition rate(Yes)
edu_attrition_yes = edu_attrition['Yes'].reset_index()

#plotting
plt.figure(figsize=(8,5))
sns.barplot(data= edu_attrition_yes, x= 'Education_level', y='Yes', palette='Set3')

#add labels
for i, val in enumerate(edu_attrition_yes['Yes']):
    plt.text(i,val+0.3, f"{val:.1f}%", ha='center', fontsize=9)

plt.title("Attrition Rate by Education Level (%)")
plt.ylabel("Attrition Rate (%)")
plt.xlabel("Education Level")
plt.show()
```

- Insight:-
-People with 'below college' attrition status has the highest attrition rate, second are people with Bachelor degree. Doctorates have the least chances of leaving the organization.

```python
#plotting attrition rate(%) by Department 

#calculating attrition rate 
dept_attrition = df.groupby('Department')['Attrition'].value_counts(normalize=True).unstack()*100

#attrition rate(Yes)
dept_attrition_yes = dept_attrition['Yes'].reset_index()

#plotting
plt.figure(figsize=(8,5))
sns.barplot(data=dept_attrition_yes, x= 'Department', y = 'Yes', palette='Set1')

#adding labels
for i, val in enumerate(dept_attrition_yes['Yes']):
    plt.text(i,val+0.3,f'{val:.1f}%', ha='center', fontsize=9)

plt.title("Attrition Rate by Department (%)")
plt.ylabel("Attrition Rate (%)")
plt.xlabel("Department")
plt.show()
```

- Insight :-
- Employees in the Sales department have the highest likelihood of leaving the organization. The HR department follows closely, with only a 1% lower attrition rate.

```python
#plotting attrition rate by Job role

#calculating attrition rate
job_role_attrition = df.groupby('JobRole')['Attrition'].value_counts(normalize=True).unstack()*100

job_role_attrition_yes = job_role_attrition['Yes'].reset_index()

#plotting
plt.figure(figsize=(12,5))
sns.barplot(data= job_role_attrition_yes, y= 'JobRole', x = 'Yes', palette='Set2')

#adding labels
for i, val in enumerate(job_role_attrition_yes['Yes']):
    plt.text(val+1,i,f'{val:.1f}%', ha='center', fontsize=9)

plt.title("Attrition Rate by Job Role(%)")
plt.xlabel("Job Role")
plt.ylabel('Attrition Rate(%)')
plt.show()
```

- Insight :-
- As observed earlier, the Sales department has the highest attrition rate. This plot further reveals that Sales Representative employees are the most likely to leave the organization.

```python
#plotting attriion rate by Income
# Create income bins
bins = [0, 5000, 10000, 15000, 20000]
labels = ['0-5K', '5K-10K', '10K-15K', '15K-20K']
df['IncomeBin'] = pd.cut(df['MonthlyIncome'], bins=bins, labels=labels)

# Calculate attrition rate by bin
income_attrition = df.groupby('IncomeBin')['Attrition'].value_counts(normalize=True).unstack()*100

income_attrition_yes = income_attrition['Yes'].reset_index()

# Plot
plt.figure(figsize=(8,5))
sns.barplot(data=income_attrition_yes, x='IncomeBin', y='Yes', palette='Set2')
plt.title("Attrition Rate by Monthly Income Range (%)")
plt.ylabel("Attrition Rate (%)")
plt.show()
```

- Insight:-
- People with the monthly income of between 0-5k has the highest chances of leaving the organization.

```python
df['WorkLifeBalance']
```

```python
#mapping of Work Life Balance column

work_life_balance_mapping = {
    1: 'Bad',
    2: 'Good',
    3: 'Better',
    4: 'Best'
}

#modifying the column
df['WorkLifeBalance'] = df['WorkLifeBalance'].map(work_life_balance_mapping)
```

```python
#plotting the attrition rate by 'WorkLifeBalance' colummn

#calculating attrition rate
work_attrition = df.groupby('WorkLifeBalance')['Attrition'].value_counts(normalize=True).unstack()*100

#attrition rate(Yes)
work_attrition_yes = work_attrition['Yes'].reset_index()

#plotting
plt.figure(figsize=(8,5))
sns.barplot(data= work_attrition_yes, x='WorkLifeBalance', y='Yes', palette='Set2')

#adding labels
for i, val in enumerate(work_attrition_yes['Yes']):
    plt.text(i,val+0.3, f'{val:.1f}%', ha='center', fontsize=9)

plt.title("Attrition Rate by Work Life Balance")
plt.ylabel("Attrition Rate (%)")
plt.show()
```

Insight:
- Its clearly visible that employees with bad work-life-balance has the highest attrition rate. Surprisingly employees in the 'Best' category has the 2nd highest chance of leaving the organization.

```python
#plotting heatmap
# Converting Attrition into numeric flag
df['Attrition_Flag'] = df['Attrition'].map({'Yes': 1, 'No': 0})

# Keep only numeric columns
numeric_df = df.select_dtypes(include=['int64', 'float64'])

# Correlation heatmap
plt.figure(figsize=(12,6))
sns.heatmap(numeric_df.corr(), cmap="coolwarm", annot=False)
plt.title("Correlation Heatmap of Numeric Features")
plt.show()

# Correlation with Attrition_Flag only
plt.figure(figsize=(6,8))
attr_corr = numeric_df.corr()[['Attrition_Flag']].sort_values(by='Attrition_Flag', ascending=False)

sns.heatmap(attr_corr, cmap="coolwarm", annot=True, cbar=False)
plt.title("Correlation of Numeric Features with Attrition")
plt.show()
```

-SUMMARY:-
- Sales employees have the highest attrition , indicating this role may involve higher stress and frequent travel.
- Employees aged between 25-35 are most likely to leave, suggesting young people are more likely to switch jobs for gaining more industrial experience.
- Male employees are more likely to leave than female workers
- Employees earning between 0-5k monthly has the highest chances of leaving the organization, indicating a possible link between lower salary ranges and higher chances of leaving.
